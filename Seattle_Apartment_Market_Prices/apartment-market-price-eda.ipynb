# Exploratory Data Analysis (EDA)
    1. Average Rent Price in Seattle
        - What's the "average" rent a person experiences?
        - Calculate the median for Tract Median Apartment Contract Rent per Unit and ...Rent per Square Foot across the whole dataset. (We use median, not mean, because rent prices are almost always skewed by a few super expensive properties)
        
    2 Market Moods / Trends
        - In any given year, what's the most likely Cost Category or Year over Year Change for a tract?
        - Get a value_counts() for the Cost Category and Year over Year Change in Rent Category columns. This will show us the distribution.
        
    3. Big Time Trend
        - How has rent in Seattle actually changed over the last 20+ years?
        - Group the data by Year and get the median Rent per Unit for each year. Then, plot this as a line chart. This is the single most important chart we can make.
        
    4. Most Least Expensive Areas
        - Where should I look if I'm rich? Where can I find a deal?
        - Filter the data for the most recent year (2024). Then, group by Community Reporting Area Name and find the median Rent per Unit. We can then sort this to find the Top 5 and Bottom 5.

```python
# Avg Rent Price in Seattle
tmacr_psf = df['Tract Median Apartment Contract Rent per Square Foot'].median()
tmacr_pu = df['Tract Median Apartment Contract Rent per Unit'].median()

print(tmacr_psf)
print(tmacr_pu)
```

```python
# Market Mood / Trends
columns_to_fix = ['Cost Category','Year over Year Change in Rent Category']
df[columns_to_fix] = df[columns_to_fix].replace('0','Unknown')

print(df['Cost Category'].value_counts())
print(df['Year over Year Change in Rent Category'].value_counts())
```

```python
# Big Time Trend
grouped_by_year = df.groupby('Year')
print(grouped_by_year['Tract Median Apartment Contract Rent per Unit'].median().sort_values(ascending=False))
```

```python
# Most Least Expensive Areas
rent_cols = ['Tract Median Apartment Contract Rent per Unit', 
             'Tract Median Apartment Contract Rent per Square Foot']
df[rent_cols] = df[rent_cols].replace(0, np.nan)

grouped_by_area = df[df['Year'] == 2024].groupby('Community Reporting Area Name')
print("Top 5 Expensive Area: ")
print(grouped_by_area['Tract Median Apartment Contract Rent per Unit'].median().sort_values(ascending=False).head(5))
print("Top 5 Least Expensive Area: ")
print(grouped_by_area['Tract Median Apartment Contract Rent per Unit'].median().sort_values(ascending=True).head(5))
```
